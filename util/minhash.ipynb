#### Outputs a file `similar_tweets.txt` in data folder

```python
import pandas as pd
from datasketch import MinHash

import util
```

```python
df = pd.read_csv('../data/covid_lies_processed.csv')
transformed = util.transform(df)
```

```python
def create_minhash(tweet: str) -> MinHash:
    mh = MinHash()
    
    for word in tweet.split():
        mh.update(word.encode('utf-8'))
        
    return mh

transformed['minhash'] = transformed.tweet.apply(lambda tweet: create_minhash(tweet))
```
